### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

```python

import pandas as pd


#Agregamos una columna que identifique las tiendas
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'


#Concatenamos los DF y generamos uno nuevo
tiendas_consolidado = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

tiendas_consolidado
```

#1. Análisis de facturación



```python
# Se genera un DF de las ventas por tienda
ventas_por_tienda = tiendas_consolidado.groupby('Tienda')['Precio'].sum().astype(int).apply(lambda x: "${:,.0f}".format(x))

# Display the result
ventas_por_tienda
```

# 2. Ventas por categoría

```python
# Agrupacion de los datos por categoria y tiendas, y se sumariza por los precios
ventas_por_categoria_y_tienda = tiendas_consolidado.groupby(["Categoría del Producto", "Tienda"])["Precio"].sum()

# Procesamiento y formateo de los precios
ventas_por_categoria_y_tienda = ventas_por_categoria_y_tienda.astype(int).apply(lambda x: "${:,.0f}".format(x))

# Visualizacion del resultado
ventas_por_categoria_y_tienda
```

# 3. Calificación promedio de la tienda


```python

# Se agrupa por Tienda y calcula el promedio de la calificacion
average_classification_by_store = tiendas_consolidado.groupby('Tienda')['Calificación'].mean()

# Visualizacion del resultado
average_classification_by_store
```

# 4. Productos más y menos vendidos

```python

# Se Agrupan los datos por tienda y producto, luego se cuentan  cada producto
product_counts = tiendas_consolidado.groupby(['Tienda', 'Producto'])['Producto'].count().reset_index(name='counts')

# Funcion que encuentra los productos mas y menos vendidos
def find_extremes(x):
    # Ordenar  y obtener el primero (máximo) y el último (mínimo)
    sorted_products = x.sort_values('counts')
    return pd.DataFrame({
        'Tienda': [x.iloc[0]['Tienda']] * 2,
        'Producto': [sorted_products.iloc[-1]['Producto'], sorted_products.iloc[0]['Producto']],
        'Type': ['Mas vendido', 'Menos vendido'],
        'Counts': [sorted_products.iloc[-1]['counts'], sorted_products.iloc[0]['counts']]
    })

# Aplicar la funcion para cada año
extremes_df = product_counts.groupby('Tienda').apply(find_extremes).reset_index(drop=True)

# Visualizacion del resultado
extremes_df
```

# 5. Envío promedio por tienda

```python
# Determinacion del costo de envio promedio por tienda
average_costo_envio_by_store = tiendas_consolidado.groupby('Tienda')['Costo de envío'].mean().apply(lambda x: "${:,.0f}".format(x))

# Visualizacion del resultado
average_costo_envio_by_store
```
